```python
import pandas as pd
```

```python
billboard=pd.read_csv('../data/billboard.csv') #reading the billboard file
billboard.head()
```

```python
billboard.isna().sum() #counting the number of missing values in each column
#if you see the output there are no missing values in first 6 columns but later the missing value count increase by 5 in wk2, 10 in wk3 so on..
```

```python
billboard.mean() # gives the mean value of each column
```

```python
billboard_Nafilledby_0=billboard.fillna(0) # filling the missing NAN values with 0
billboard_Nafilledby_0.isna().sum() # this would result in the count of 0's to be=0 since we filled the null values with 0
```

```python
billboard_Nafilledby_mean=billboard.fillna(billboard.mean()) # filling the missing NAN values with mean value of their respective columns
billboard_Nafilledby_mean.isna().sum() # this would result in the count of 0's to be=0 since we filled the null values with mean
```

```python
print(billboard.median())
billboard_Nafilledby_median=billboard.fillna(billboard.median()) # filling the missing NAN values with median value of their respective columns
billboard_Nafilledby_median.isna().sum() # this would result in the count of 0's to be=0 since we filled the null values with median
```

```python
print(billboard.mode())
k=billboard.loc[:,"wk2"].mode()
billboard.fillna({'wk2':k[0]}) #	Here if you observe 17th row wk2 is having NaN value.
#now we shall fill that with mode for that corresponding column.
```
